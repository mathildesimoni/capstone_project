```python
import pandas as pd
import time
import random
import os
from os.path import exists
from countrygroups import EUROPEAN_UNION
import pyreadr
import numpy as np

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

```python
result = pyreadr.read_r('dataverse_files/firstNameCounts.RData')
# print(result.keys()) # let's check what objects we got, now only df1 is listed
df_first = result["firstNameCounts"] # extract the pandas data frame for object df1

df_first['first_name'] = df_first['first_name'].str.lower()
df_first = df_first.dropna()
df_first = df_first[df_first['first_name'].str.len() > 1]

firsts = df_first['first_name']
fs = []
for f in firsts:
    tmp = f.split(" ")[-1]
    if tmp[0] == "'":
        tmp = tmp[1:]
    fs.append(tmp)
    
df_first['first_name'] = fs
df_first = df_first[df_first['first_name'].str.len() > 1]

display(df_first)
```

```python
print(len(df_first['first_name']))
f = np.unique(np.array(list(df_first['first_name'])))
print(len(f))

df_first = df_first.groupby("first_name")["p_whi_first", "p_bla_first", "p_his_first", "p_asi_first", "p_oth_first"].apply(lambda x : x.astype(int).sum()).reset_index()
display(df_first)
```

```python
print(list(df_first.iloc[0]))
print("'")
```

```python
names = []
races = []
maxes = []
for i in range(len(df_first)):
    arr = np.array(df_first.iloc[i])
    i_max = np.argmax(arr[1:])
    if i_max == 0:
        races.append("White")
        names.append(arr[0])
        maxes.append(arr[1:][i_max])
    elif i_max == 1:
        races.append("Black")
        names.append(arr[0])
        maxes.append(arr[1:][i_max])
    elif i_max == 2:
        races.append("Hispanic")
        names.append(arr[0])
        maxes.append(arr[1:][i_max])
    elif i_max == 3:
        races.append("Asian")
        names.append(arr[0])
        maxes.append(arr[1:][i_max])

df_final_first = pd.DataFrame()
df_final_first['FirstName'] = names
df_final_first['Race'] = races
display(df_final_first)
```

```python
df_final_first1 = df_final_first.drop_duplicates()
display(df_final_first1)
```

```python
result = pyreadr.read_r('dataverse_files/lastNameCounts.RData')
print(result.keys()) # let's check what objects we got, now only df1 is listed
df_last = result["lastNameCounts"] # extract the pandas data frame for object df1

df_last['last_name'] = df_last['last_name'].str.lower()
df_last = df_last.dropna()
df_last = df_last[df_last['last_name'].str.len() > 1]

lasts = df_last['last_name']
fs = []
for f in lasts:
    tmp = f.split(" ")[-1]
    if tmp[0] == "'":
        tmp = tmp[1:]
    fs.append(tmp)
    
df_last['last_name'] = fs
df_last = df_last[df_last['last_name'].str.len() > 1]

display(df_last)
```

```python
print(len(df_last['last_name']))
f = np.unique(np.array(list(df_last['last_name'])))
print(len(f))

df_last = df_last.groupby("last_name")["p_whi_last", "p_bla_last", "p_his_last", "p_asi_last", "p_oth_last"].apply(lambda x : x.astype(int).sum()).reset_index()
display(df_last)
```

```python
names = []
races = []
for i in range(len(df_last)):
    arr = np.array(df_last.iloc[i])
    i_max = np.argmax(arr[1:])
    if i_max == 0:
        races.append("White")
        names.append(arr[0])
    elif i_max == 1:
        races.append("Black")
        names.append(arr[0])
    elif i_max == 2:
        races.append("Hispanic")
        names.append(arr[0])
    elif i_max == 3:
        races.append("Asian")
        names.append(arr[0])

df_final_last = pd.DataFrame()
df_final_last['LastName'] = names
df_final_last['Race'] = races
display(df_final_last)
```

```python
df_final_first.to_csv("voters_first_names.csv", index = False)
df_final_last.to_csv("voters_last_names.csv", index = False)
```
