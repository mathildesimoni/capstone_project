```python
import pandas as pd
import time
import random
import os
from os.path import exists
import pyreadr
import numpy as np
from itertools import product

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

```python
thresh = 0.9
```

```python
result = pyreadr.read_r('dataverse_files/firstNameCounts.RData')
# print(result.keys()) # let's check what objects we got, now only df1 is listed
df_first = result["firstNameCounts"] # extract the pandas data frame for object df1

df_first['first_name'] = df_first['first_name'].str.lower()
df_first = df_first.dropna()
df_first = df_first[df_first['first_name'].str.len() > 1]

firsts = df_first['first_name']
fs = []
for f in firsts:
    tmp = f.split(" ")[-1]
    if tmp[0] == "'":
        tmp = tmp[1:]
    fs.append(tmp)
    
df_first['first_name'] = fs
df_first = df_first[df_first['first_name'].str.len() > 1]

display(df_first)

# df_first = df_first.groupby("first_name")["p_whi_first", "p_bla_first", "p_his_first", "p_asi_first", "p_oth_first"].apply(lambda x : x.astype(int).sum()).reset_index()
# display(df_first)
```

```python
# df_first = df_first.groupby("first_name")["p_whi_first", "p_bla_first", "p_his_first", "p_asi_first", "p_oth_first"].apply(lambda x : x.astype(int).sum()).reset_index()
# display(df_first)
# df_first.to_csv("tmp_df_first.csv", index = False)

df_first = pd.read_csv("tmp_df_first.csv")
display(df_first)
```

```python
df_first['total'] = df_first["p_whi_first"] + df_first["p_bla_first"] + df_first["p_his_first"] + df_first["p_asi_first"] + df_first["p_oth_first"]
display(df_first)
```

```python
names = []
races = []
maxes = []
for i in range(len(df_first)):
    if i %100000 == 0:
        print(i)
    arr = np.array(df_first.iloc[i])
    max_p = max(arr[1:-1])
    p = max_p/arr[-1]
    if p >= thresh:
        i_max = np.argmax(arr[1:-1])
        if i_max == 0:
            races.append("White")
            names.append(arr[0])
            maxes.append(arr[1:-1][i_max])
        elif i_max == 1:
            races.append("Black")
            names.append(arr[0])
            maxes.append(arr[1:-1][i_max])
        elif i_max == 2:
            races.append("Hispanic")
            names.append(arr[0])
            maxes.append(arr[1:-1][i_max])
        elif i_max == 3:
            races.append("Asian")
            names.append(arr[0])
            maxes.append(arr[1:-1][i_max])

df_final_first = pd.DataFrame()
df_final_first['FirstName'] = names
df_final_first['Race'] = races
display(df_final_first)
```
